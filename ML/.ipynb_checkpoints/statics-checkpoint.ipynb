```python
def plot_normal_curve(mu=0, sigma = 1, x_left=-10,x_right=10):
    x = np.linspace(x_left,x_right,100)
    a = 1/(sigma*np.sqrt(2*np.pi))
    b= -1/2*((x-mu)/sigma)**2
    f_x = a* np.exp(b)
    plt.plot(x,f_x)
```

```python
plot_normal_curve() #standard normal = neam = 0 ; std = 1
```

## probability distribution 

```python
p_df = pd.read_csv("c4_epa_air_quality.csv", index_col = 0)
```

```python
p_df['aqi'].hist();
```

```python
mean_aqi = df['aqi'].mean()
std_aqi = df['aqi'].std()
print(mean_aqi,std_aqi)
```

```python
#1st standard deviation --. 68% or more
lower_limit = mean_aqi - 1 * std_aqi
upper_limit = mean_aqi +1 * std_aqi

print(lower_limit,upper_limit)
```

```python
((df['aqi'] >= lower_limit ) & (df['aqi'] <= upper_limit)).mean()*100
```

```python
#2nd standard deviation
lower_limit = mean_aqi - 2 * std_aqi
upper_limit = mean_aqi +2 * std_aqi

print(lower_limit,upper_limit)
((df['aqi'] >= lower_limit ) & (df['aqi'] <= upper_limit)).mean()*100
```

```python
#3rd standart deviaton

lower_limit = mean_aqi - 3 * std_aqi
upper_limit = mean_aqi +3 * std_aqi

print(lower_limit,upper_limit)
((df['aqi'] >= lower_limit ) & (df['aqi'] <= upper_limit)).mean()*100
```

```python
df['Z_score'] = stats.zscore(df['aqi'],ddof = 1)
```

```python
df.head()
```

```python
df[(df['Z_score']>3) | (df["Z_score"]< -3)]
```

## Confidence intervals

```python
df.head()
```

```python
df.describe(include = 'all')
```

```python
states = ['California','Florida','Michigan','Ohio','Pennsylvania','Texas']
con_interval = df[df['state_name'].isin(states)]
con_interval.groupby(['state_name']).agg({'aqi':'mean','state_name':'count'})
```

```python
import seaborn as sns
sns.boxplot(x=con_interval['state_name'],y=con_interval['aqi'])
```

```python
aqi_ca = df[df['state_name']=='California']
sample_mean = aqi_ca['aqi'].mean()
sample_mean
```

```python
confidence_level = 0.95 
confidence_level
```

```python
#begin by identifying the z associated with your chosen confidence level
z_value = 1.95

#next, calculate your standard error
standard_error = aqi_ca['aqi'].std()/np.sqrt(aqi_ca.shape[0])
print("Standard erroer")
print(standard_error)

#lastly 
margin_of_error = standard_error * z_value
print("margin_of_error")
print(margin_of_error)
```

```python
upper_ci_limit = sample_mean + margin_of_error
lower_ci_limit = sample_mean - margin_of_error
(lower_ci_limit,upper_ci_limit)
```
